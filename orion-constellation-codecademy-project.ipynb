## Project: Visualizing the Orion Constellation

## 1. Set-Up

```python
%matplotlib notebook
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## 2. Get familiar with real data


```python
# Orion
x = [-0.41, 0.57, 0.07, 0.00, -0.29, -0.32,-0.50,-0.23, -0.23]
y = [4.12, 7.71, 2.36, 9.10, 13.35, 8.13, 7.19, 13.25,13.43]
z = [2.06, 0.84, 1.56, 2.07, 2.36, 1.72, 0.66, 1.25,1.38]
```

## 3. Create a 2D Visualization

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.scatter(x,y, color='y', marker='*')
plt.title('The Orion Constellation in 2D')
plt.xlabel('x coordinates')
plt.ylabel('y coordinates')
ax.set_xticks(x)

yticks = [max(y)]         
for i in sorted(y):
  if y[-1] - i > 0.18:      
    yticks.append(i) 
    
ax.set_yticks(yticks)
ax.set_xticklabels(x, fontsize=8, rotation='45')
ax.set_yticklabels(yticks, fontsize=8)
ax.set_facecolor('xkcd:dark blue')
plt.show()
```

## 4. Create a 3D Visualization

```python
fig_3d = plt.figure()
constellation3d=fig_3d.add_subplot(1,1,1, projection ='3d')
constellation3d.scatter(x,y,z, color='yellow',marker='*', s=60)
plt.title('The Orion Constellation in 3D')
constellation3d.set_xlabel('x coordinates')
constellation3d.set_ylabel('y coordinates')
constellation3d.set_zlabel('z coordinates')
constellation3d.w_xaxis.set_pane_color((0, 0, 0.3, 1))
constellation3d.w_yaxis.set_pane_color((0, 0, 0.3, 1))
constellation3d.w_zaxis.set_pane_color((0, 0, 0.3, 1))
plt.show()
```
